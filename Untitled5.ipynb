```python
### Importing Required Libraries

import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D,Flatten,Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
df = pd.read_csv("/content/facial-expression/fer2013.csv")
```

```python
### Checking the image

plt.imshow(np.array(df.pixels.loc[0].split(" ")).reshape(48,48).astype("float"))
```

```python
### Checking multiple images to understand its labels

fig = plt.figure(1, (14, 14))
k = 0
for i in range(min(49, len(df))):
    px = df.pixels.iloc[k]
    px = np.array(px.split(' ')).reshape(48, 48).astype('float32')
    k += 1
    ax = plt.subplot(7, 7, k)
    ax.imshow(px, cmap='gray')  # Use the 'gray' colormap for grayscale images
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(f'Emotion: {df.emotion.iloc[i]}')
    plt.tight_layout()

plt.show()
```

```python
### Now Assigning lables based on the image previous images

label_to_text = {0:'anger', 1:'disgust', 2:'fear', 3:'happiness', 4: 'sadness', 5: 'surprise', 6: 'neutral'}
```

```python
df['emotion_label'] = df['emotion'].map(label_to_text)
df.head()
```

```python
df.isnull().sum()
```

```python
pixels = df.pixels.apply(lambda x: np.array(x.split(" ")).reshape(48, 48,1).astype("float32"))
```

```python
pixels.shape
```

```python
pixels_3d = np.stack(pixels, axis=0)
```

```python
pixels_3d.shape
```

```python
labels = df.emotion.values
labels
```

```python
### Spliting the dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pixels_3d, labels, test_size=0.33, random_state=42)
```

```python
# Checking the shape of the training and test sets

print('Shape of the training input data:', X_train.shape)
print('Shape of the training output data:', y_train.shape)
print('Shape of the test input data:', X_test.shape)
print('Shape of the test output data:', y_test.shape)
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), padding="valid", activation="relu", input_shape=(48, 48, 1)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding="valid"))
model.add(Dropout(0.3))

model.add(Conv2D(64, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding="valid"))
model.add(Dropout(0.3))

model.add(Conv2D(128, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Conv2D(128, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Flatten())

model.add(Dense(128, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(64, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.1))

model.add(Dense(7, activation="softmax"))

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.summary()
```

```python
### Applying early stopping

early_stopping = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping]
)
```

**Applying Transfer Learning Technique**

```python
from keras.applications import ResNet50
```

```python
conv_base = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
conv_base.trainable = True
set_trainable = False
for layer in conv_base.layers:
    if layer.name == 'conv5_block1_0_conv':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```python
model = Sequential()
model.add(conv_base)
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(7, activation='softmax'))
```

```python
from keras.optimizers import Adam
model.compile(optimizer=Adam(lr=1e-5), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```
